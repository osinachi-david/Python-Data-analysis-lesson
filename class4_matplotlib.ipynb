```python
#Create some data 
years = [1970, 1980, 1990, 2000, 2010, 2020, 2030]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
#create a line chart 
#years on the x-axis and gdp on the y-axis 

plt.plot(years, gdp)
plt.show()
```

```python
#improve the chart 
# customize the colors 
plt.plot(years, gdp, color='#00000')
```

```python
fig, ax = plt.subplots()
ax.set_facecolor((0.1, 0.2, 0.5))
plt.title('Nominal GDP') # add title 
plt.ylabel("billions of dollars $")
# plt.xlabel('Years')
ax.plot(years, gdp, color=(1.0, 1.0, 1.0), marker='X', linestyle='dashed')
```
