# PIVOT TABLES 

The pivot table takes a simple column-wise data as input and groups the entries into a 2-D table that provides a multi-dimensional summary of the data.

This tool enables users to automatically count, sort, total, or average the data stored in one table 

```python
# environment setup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 

print("Ready")
```

```python
data = sns.load_dataset('exercise')
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.info()
```

```python
pd.pivot_table(diamonds, index=["cut"])
```

```python
pi1 = pd.read_excel('datasets/pi1.xlsx')
```

```python
pi2 = pd.read_excel("datasets/pi2.xlsx")
```

```python
pi2.head(2)
```

```python
pvs = pi2.pivot_table(index=["Salesperson"], values=["Total Cost"], aggfunc="sum")
```

```python
# Total sales per sales person
pvs
```

```python
pd.pivot_table(pi2, index=["Salesperson"], values=["Product"], aggfunc="count")
```

```python
pd.pivot_table(pi2, index=["Salesperson", "Customer"], aggfunc="min")
```

```python
pd.pivot_table(pi2, index=["Salesperson", "Customer", "Region"], aggfunc="sum")
```

```python
pd.pivot_table(pi2, index=["Salesperson", "Customer", "Region"], values=["Total Cost"], aggfunc=[np.mean, len])
```

```python
# change the labels for the aggregate functions 
pd.pivot_table(pi2, index=["Salesperson", "Customer", "Region"], values=["Total Cost", "No.Items"], aggfunc={"Total Cost":np.sum, "No.Items": len})
```

### Columns and Values

```python
# columns provide additional ways to segment your values 
# note that aggregate functions are applied to the values 
```

```python
pd.pivot_table(pi2, index=["Salesperson", "Customer"], values="Total Cost", columns="Product", aggfunc="sum")
```

```python
pd.pivot_table(pi2, index=["Salesperson", "Customer"], values=["Total Cost", "No.Items"], columns="Product", aggfunc="sum", fill_value=0)
```

```python
# Region level
pd.pivot_table(pi2, index=["Region","Product", "Customer"], values=["Total Cost", "No.Items"], aggfunc="sum", fill_value=0)
```

```python
# store your pivot table in a variable 

table = pd.pivot_table(pi2, index=["Region","Product"], values=["No.Items"], aggfunc="sum", fill_value=0)
```

```python
table.plot(kind="bar")
```

```python
table1=pd.pivot_table(pi2, index=["Salesperson"], values=["No.Items"],columns="Product", aggfunc="sum",fill_value=0)
```

```python
# table1.plot(kind="bar"),
sns.set()
table1.plot(kind="bar")
```

```python
table = pi2.pivot_table(index=["Salesperson", "Customer"], values="No.Items", aggfunc="count", fill_value=0)
```

```python
# get the 1st 100 rows but shuffle the dataset first 
df = diamonds.sample(frac=1).head(100)
```

```python
df.head(3)
```

```python
sns.lmplot(data=df, x='price', y='carat', scatter_kws={'color':'red'})
```

```python
# higher order trend lines 
sns.lmplot(data=df, x='price', y='carat', order=2, scatter_kws={'color':'red'})
```

```python
# plotting without the trend line
# higher order trend lines 
sns.lmplot(data=df, fit_reg=False, x='price', y='carat', scatter_kws={'color':'purple'})
```

```python
# adding more args 
# hue = cut 
sns.lmplot(data=df, fit_reg=False, x='price', y='carat', hue="cut",)
```

```python
# regplot
sns.regplot(data=df, x='price', y='carat')
```

```python
df.head(2)
```

```python
price = df["price"]
carat = df["carat"]
```

```python
x = price
y = carat
```

```python
from scipy import stats
```

```python
stats.linregress(x,y)
```

```python
#destructure 
slope, intercept, r, p, stderr = stats.linregress(x,y)

#create a func
def myFunc(x):
    return slope * x + intercept

# run each vaue of x through the function above
# this will generate new values for y-axis 

new_y = list(map(myFunc, x))

plt.scatter(x,y)

plt.plot(x, new_y)
```
