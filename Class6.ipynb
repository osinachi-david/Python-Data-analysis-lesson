```python
# dropna(parameter)
# how = 'any'
# how = 'all'

emp.dropna(how='any') # default 

emp.dropna(how='all').head(2) # if all the values in a row are null 
# the row will be dropped 
```

```python
# with columns 

emp.dropna(axis=1, how='any') # if 1 null value is found col will be dropped 

emp.dropna(axis=1, how='all').tail(3) # all values must be null
```

```python
# dropna(parameter)
# thresh = no_of_null_values 
# thresh = 3

emp.dropna(thresh=2, axis=1).head(3)
```

## replacing null values 

### your decision to replace null values is dependent on the dataset you're working on 
- replace Null with 0
- replace with the sample mean
- replace with the closest value 

```python
s = pd.Series([1,2,3,4,np.nan,5,6,np.nan,3])
```

```python
#replace with 0
s.fillna(0)
```

```python
emp.fillna(0).head(3)
```

```python
s.fillna(s.mean())
```

```python
emp.fillna(emp.mean()).head(3)
```

```python
# replace with the closest value 
# fowardfill and # backwardfill
s.fillna(method='ffill')
s.ffill()
```

```python
# backwardfill
s.fillna(method='bfill')
s.backfill()
```

```python
emp.bfill()
```
